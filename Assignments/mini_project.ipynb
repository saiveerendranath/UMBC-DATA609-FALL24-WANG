## Step 4 - Categorize Each Country

Create a new column "Group". If a country is one of the seven G7 countried, the value of "Group" would be "G7", if a country of one of the five BRICS countries, the value of "Group" would be "BRICS". Otherwise, the value of "Group" would be "REST of WORLD".

```python
countries_brics = ["Brazil", "Russian Federation", "India", "China", "South Africa"]
countries_g7 = ["Canada", "France", "Germany", "Italy", "Japan", "United Kingdom", "United States"]
```

```python
def country_group(country):
  if country in countries_brics:
    return "BRICS"
  elif country in countries_g7:
    return "G7"
  else:
    return "REST of WORLD"
```

```python
g7brics_df['Group'] = g7brics_df['Country Name'].apply(country_group)
g7brics_df.sample(5)
```

## Step 5 - Aggregate Data

Aggregate the following columns at the "Group" level for each year:

- For GDP, using "sum" for aggregation.
- For GDP Per Capita, use "mean" for aggregation.
- Check your calculation with the reference site.

```python
g7brics_group = g7brics_df.groupby(['Year','Group']).aggregate({
    'GDP ($Billions)': 'sum',
    'GDP Per Capita ($)': 'mean'
})

g7brics_group
```

```python
g7brics_group = g7brics_group.reset_index()
g7brics_group
```

## Step 6 - Calculate Percentage Share of GDP

For each year, calculate the percentage share of each group's GDP. Check your calculation with the reference site.

```python
def percentage_share_gdp(row):
  gdp_usd_total = g7brics_group[g7brics_group["Year"] == row["Year"]]['GDP ($Billions)'].sum()
  row['GDP ($Billions)' + " (%)"] = round(100 * row['GDP ($Billions)'] / gdp_usd_total, 1)
  return row
```

```python
g7brics_group_gdp_per = g7brics_group.apply(percentage_share_gdp,axis=1)
g7brics_group_gdp_per.sample(5)
```

## Step 7 - Plot the GDP Percentage Share Over Time

```python
fig = px.bar(
    g7brics_group_gdp_per,
    x='Year',
    y = 'GDP ($Billions) (%)',
    color = 'Group',
    title="GDP Percentage Share Over Time",
    text='GDP ($Billions) (%)'
)
```

```python

fig.update_layout(
    xaxis_title="Year",
    yaxis_title="GDP Percentage Share",
    legend_title="Group",
    title = {
        'xanchor': 'center',
        'yanchor': 'top',
        'x': 0.5,
    }
).show()
```
