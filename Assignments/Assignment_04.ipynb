## HW Question 2

Find out if there is a relationship between wealth and health

Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.

```python
df3['Year'].unique()
```

```python
sample_df = df3[df3['Year'] == 2021]
sample_df
```

```python
import plotly.express as px

scatter_fig = px.scatter(sample_df, x = 'Life expectancy at birth, total (years)', y = 'GDP per capita (current US$)')

scatter_fig.show()
```

```python
import plotly.express as px

scatter_fig = px.scatter(sample_df, y = 'Life expectancy at birth, total (years)', x = 'Income Group')

scatter_fig.show()
```


Observation from the above scatter plots -

Observation from 1st Scatter Plot:
*   For <50k GDP we can find more density below <75, which determines.
*   For Life Expectency >75 we can see there are almost equal density among GDP <50k and >50k.

Observation from 2nd Scatter Plot:

*   Low Income, Lower Middle Income has less Life expectency when compared to Upper Middle Income and High Income.
*   Order of Life expectency goes as Low Income < Lower Middle Income ~< Upper Middle Income < High Income.

From the above observation we can see that Life Expectency is directly proportional to Income Group.






## Homework Question 3

Select a year of data. calculate total population of each region and make a pie chart to display the distribution of total population of all regions.

```python
sample_data = df3[df3['Year'] == 2020]
sample_data
```

```python
print(sample_data['Population, total'].dtype)
```

```python
sample_data['Region'].unique()
```

```python
population_per_region_df = sample_data.groupby('Region')['Population, total'].sum().reset_index()
population_per_region_df
```

```python
pie_fig = px.pie(population_per_region_df, values = 'Population, total', names = 'Region')
pie_fig.show()
```
